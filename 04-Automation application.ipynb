# Automation can fit in ISTM, IAM, Enterprise IT

# ISTM Automation
1. Incident Management
   * auto creating tickets from monitoring systems/alerts
   * auto assign to the right team & right person
   * Trigger auto-remedies - restart the services, clear cache
2. Service Requests:
   * Self service protals worflow automation
3. Change Management:
   * standard changes (low risk) can be automated
   * automated deployment pipelines for production deployment reducing the human intevention
4. Problem Management:
   * Automating recurring issues with data patterns

---
# IAM Automation
1. User lifecycle management
    * joiner - auto creation of accounts
    * mover - update operation based on role change
    * leaver - revoking the access when employees quit
    * on hold - employee is taking a break - maternity, study break, medical break, etc..
2. Access Request:
   * automated approval worksflows
   * policy-based access granting
3. Role-Based Access Control
   * Atomated permissions assignments based on roles associated with job
4. Access Reviews and Compliance
   * periodic access to certification campaings or courses
   * automated audit logs
   * automated reports generation

# IT operations
1. Infrastructure automation
   * creating vms
   * creating vpc, subnet, vpn, firewall rules
   * creating incoming and outgoing request/response rules
   * creating networks
   * creating any kind of cloud resources
2. DevOps
   * Code management - version control workflows --> PR based workflows, build, testing, packing
   * Deployment activities - dev env, test env, pre prod env, prod
   * monitoring - alert generation, tickets management
3. Security Operations
   * automatically updating the os/software packages
   * bug fixes, security patches, etc..
   * threat detection and auto respose (SOAR)
4. Business Process Integration
   * HR related activities
   * finance related activities
   * IT related activities
   * any kind system activities
