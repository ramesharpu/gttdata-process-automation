# Selenium

---
Selenium Web Automation tool, it is used to autmate interactions on the web pased interfaces by mimicing the actions/behaviors like clicking, typing, navigation, selecting, etc..

---
# Strengths
* It is opensource tool ==> free of cost to use even in the production enviroment
* It supports following
  - multiple operating system => windows, linux, mac
  - multiple browsers => chrome, firefox, edge, opera, safari, etc..
  - multiple languages => java, python, c#, ruby, golang, etc..
  - multiple architectures => 64 bit, 32 bit
  - multiple web frameworks => mvc, soa, component, etc..
  - multiple older technologies ==> jsp, servlets, ejbs, etc
  - multiple latest technologies => react, node, angular, vuejs, nextjs, etc
* It has defined the web standards ==> wedriver 2.0 standards
* Automate browser based administative activities/tasks(eg. AD self-service poratls, admin consoles)
* Userful when no API, CLI or direct integration is available
* Can bridge gaps between legacy identity management systems and modern workflows

---
# Best Fit use cases
* Legacy AD management portals without API support
* Third party identiy tools with web only interfaces
* One time migration or transitional automation like bulk updates via UI
* Stopgap solution while proper automation (powershell/api) is being developed

---
# Limitations
* Can be fragile if the ui changes are very frequent
* Requires good coding knowledge to write and maintain the scripts
* Slower than the powershell or api based automation
* Poor auditabiltiy, but can be completely overcomed using the reporting option


---
# Design considerstions
* Desing with Data driven, keyword driven or hybrid framework
* Always try to use the standard data serialization format like csv, yaml, json
* If the desing pref requires to use poweshell or api scripts, you can always integrate for better efficiecy and trackability
* Implement strong error handling and logging for traceability
* Montor for UI changes that may break the automation
* Secure credentials (avoid hard coding, use secrets or vaults)
* include the fallback or manual override process
