# Automation in IT Operations
IT operations automation focuses on managing
1. infrastructure
2. systems
3. services

it is going to be minimal manual efforts

**Key Areas:**
1. Provisioning and Configuring the infra
   * Automatically setup the servers, db's, environments, managing with Load Balancers
   * scripts/tools to configure it according to business need and maintain the consistency
2. Conitunes Monitoring & incident Management & Quick Responses
   * systems should continuesly monitor the performance and health of the infra
   * alerts triggered should be automatic and actions needs to be performed here on the go(restart service, scale resources)
3. Deployment and Updates
   * Continues Integration (CI), Continues Testing (CT), Continues Deoployment (CD)
   * Almost zero downtime, and Almost zero errors
4. Backup and Recovery
   * Schedules backups runs without the human intervention
   * Automated recovery processes to reduce any kind of downtime

# Automation in Identity Management
Identity management automation focused on controlling who has access to what and ensuring its always correct. Authentication and Authorization needs to be managed without any errors

**Key Areas:**
1. User Provisioning (Onboarding)
   * Automatically creating accounts when a new employee join
   * Assigning roles, permissions , access rights
   * eg: email, system access, application permissions
2. De-provisioning (Offboarding)
   * Instantly revoke access when an employee leaves
   * Prevent security risks
   * eg: employee exits ==> all the accounts will be disabled immediately
3. Role Based Access Control (RBAC):
   * Access is assigned based on roles (eg. developer, tester, manager, hr, admin, security, etc..)
   * Automation to ensure consistent permission are assigned
4. Access Request and Approvals
   * Automated workflows for requesting access
   * Manager approvals should happen digitally
5. Audit & Compliance
   * All kinds of logs access and changes required
   * Backups need to be done automatically
   * Audit reports needs to generated automatically
6. Benefits in identity management
   * Stronger Security in place
   * Reduced risk of unauthorization access
   * Faster onboarding/ offboarding
   * Easier compliance adherence with regulation
