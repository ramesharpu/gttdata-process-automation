# What makes a process automateable

Whent he characteristics allow a software or machine to execute a task with minimal human judgement ==> process is automatable

### Key features which makes a process suitable for automation:
**1. Rule based and well defined**
* If the process follows clear, unambiguous rules (if x then y), its very easy to automate.
* If the judgement is not proper or clearly defined, automating it is difficult

**2. Common task or Repetitive task or High Volume task**
* Any common task
* Any Repetitive task
* Any high volume

**3. Structed Inputs and Outputs**
* Valid data, invalid data
* form and data fields ==> it should be validated according

**4. Stable and Mature Process**
* If the product features are keep on changing
* Requiring maintainance work
* Stability is not there
* If there is variations with planned and actual activity

**5. Low Exception Rate**
* Process with few edge cases or exception are easier to automate
* if the edges/special case are vast and requires human input/interventional ==> automation is going be less effective

**6. Digital and Accessible Data**
* Any digital activity is easy to automate
* If paper work is involved, its very difficult to automate

**7. Time Sensitive or Error Prone**
* Process where speed and accuracy matters(eg. complicance checks, calculations) ==> automation can be a great fit
* Machines can perform the activity very face and accurate
* Humans might have errors

**8. Clear inputs ==> Output predictable**
* If outcome is predictable consistently with given input, automation is very easy, feasible, flexible
* If not, automation will unpredictable or less creative
* Creative task ==> automation is very difficult
