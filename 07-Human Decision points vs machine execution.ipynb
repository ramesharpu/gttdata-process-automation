# Human Decision points vs machine execution

Understanding the split between humans decision points and machine execution is key to designing effective automation.
The goal of automation is not to remove humans - its about placing them where they add the most value

---
# Human Decisions Points

Judgement, Context, Responsibility

**Why humans are essential:**
1. Complex Decisions
2. Handle Situations with ambiguity or incomplete data
3. Trade-offs which requiring experience
4. Exceptions or Edge Cases
5. Rare or Unexepected behaviors
6. Uses Cases that doesnot fit in predefined rules
7. Decision involving policies, legal risks or ethics
8. Sensitive information handling
9. Strategy Thinking
10. Planning, optimization, long-term decisions


---
# Machine Execution
Machines (automation) are strucutred to handle repeatable, rule based and boring jobs
1. Repetative task ==> huge volume based work
2. Rule base decision ==> if developer --> assign the git + server access
3. Speed and Scaling ==> Parallel execution, monitoring thousands of systems and alerting in real time
4. Consitency ==> apply security policies uniformly
