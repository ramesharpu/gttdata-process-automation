# Overview of Automation Tools Landscape

There are varieties of automation tools, based on the needs for our productivity we can choose the relevant one and do the automation

# Categories of Automation Tools

### 1. Programming/Scripting-based automation
* Use programming or scripting language (java, python, shell, bash, powershell)
* Highly flexible and customization
* Examples - Shell, bash, selenium, appium, restassured, unirest, ansible, chef, puppet, etc...

Best for:
* System administration
* DevOps
* Infrastructure automation
* Data processing

**Pros:**
* Maximum control
* No licensing cost (open source)
  
**Cons:**
* Required skilled developers
* Harder to maintain at scale

--- 
### 2. Workflow-based automation
* Visuale workflow with predefined steps and logics
* Integrate with multiple systems via APIs
* Examples - Apache Airflow, Camunda, etc..

Best for:
* Business processes
* approvals
* orchestration

**Pros:**
* Easier to visualize the process
* Better for cross-team collaboration
  
**Cons:**
* Limited flexibility
* Limited coding
* Can become complexx for advanced logic

---
### 3. Low code or No code automation
* Drag and drop interfaces
* Record and playback
* examples - Microsoft Power Automate, zapier, Selenium IDE

Best for:
* Business users
* Quick automation
* Integrations

**Pros:**
* Fast Implementation
* Accesible for non-developers
  
**Cons:**
* Limited Customization
* Scaling is difficult
* Governance challenging

---
### 4. ISTM integrated Tools
* Built into IT service Management platforms
* Automate incident, change and request workflows
* Example - ServiceNow, BMC Helix

Best for:
* IT operations and service management

**Pros:**
* NAtive integration with IT process
* Strong governance and audit capabilities
  
**Cons:**
* Expensive
* Customization can complex or sometimes even not possible
