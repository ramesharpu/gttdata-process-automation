# Tools Consideration

---
# 1. Scripting Tools ==> Powershell

Poweshell is a natual fit for AD automation. Because powershell is tightly integrated with Active Directory, making it the defualt automation interface for administrators

**Key Advantages:**
* Native AD module support
* Direct ovject manipulation (user, groups, OUs)
* Scriptable and repeatable operation
* Remove execution at scale
* Strongly supports for bulk operations and pipelines

---
# Sample code

**User Creation:**
```powershell
New-ADUser -Name "John Rambo" -GivenName "John" -Surname "Rambo" -SamAccountName "jrambo" -UsePrincipalName "jrambo@company.com"
-Department "Finance" -Enabled $true
```

**Group Association:**
```powershell
Add-ADGroupMemeber -Identity "Finance" -Members "jrambo"
```

---
# Best fit use cases
1. Joiner-Mover-Leaver (JML) automation
2. Bulk user provisioning
3. Group-based access control updates
4. AD cleanup activities like stale accounts, inactive users
5. Scheduled tasks and background jobs

--- 
# When to use it
* speed and efficiency
* consistency
* flexibility
* integration

---
# Limitations
1. No Native Governance
   * no built-in approvals or access certification
   * scripts can bypass the policy if not controlled
2. Error handling complexity
   * poorly written scripts ==> can cause large scale issues very quicy
3. Security Risks
   * Hardcoded credentials
   * Over privileged execution contexts
4. Maintainability
   * Scripts can become hard to manage without standards
