# Active Directory Platform

---
# Strengths of AD based automation

Native identity and access management (IAM)
* users, group ==> creation and management is
* computers/vms are connected internally to AD server
* Authentication (kerberos, LDAP)
* Authorization via group membership and policies

automation is fast, reliable and close to enforcement point

---
# Direct object level operation
1. Create/ modify/ delete ==> user and group level
2. Assiging group membershipt instantly
3. Apply policies via organization units

low latency, high precision automation without needing multiple intermediates

---
# Best user cases for automation
1. user management
2. group management
3. associating users with groups
4. ou placement for policy inheritance
5. provisioning and deprovisioning
6. disablement of users/groups
7. triggering downstream provision ==> apps relying on AD groups



---
# Limitations
1. Limited workflow and approval capabilites. AD does not natively support the following
   * multi-step approval workflows
   * business friendly request
   * policy based desicion engines
2. Higher risk without governance layer
   * Misconfigured scripts can grant excessive or lesser access
   * no build in seperation of duties
   * limited contextual awareness


---
# Design Considerations
1. Security Boundaries
   * use tiered administrations (tier 0/1/2)
   * restrict who/what can automate changes in AD
   * isolated critical objects (eg., domain admins, previleged groups)
   * Automation should respect privilege boundaries and not bypass them
2. Change control and auditability
   * Log all automation actions (who/what/when/why)
   * keep backup of these log files
   * integrate with siem or audit systems
   * use controlled pipeline and not adhoc scripts
   * avoid event based automations and trigger manually as much as possilbe
3. Keep track of everything
   * Traceability of each activity
   * reversible operation (wherever it is possible)
   * policy algined workflows
