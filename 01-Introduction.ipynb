# Automation

Is the use of technology to perform the tasks with minimal human intervention

**The goal is usually:**
* increase efficiency
* reduce errors
* save time
* improve the quality of deliverables
* improve the process
* imporve the performance
* improve teh consistency

Different task require differnt automation. So, broadly its classified into two categories
1. IT Automation
2. Process Automation

## IT Automation:
Developers, Testers, Devops ==> they go ahead automating their activities like
* code management ==> version control and releases
* code ==> built, tested and pacakged ==> Continues Integration & Continues Testing
* Infra needs to created and managed ==> Configuration management
* Deployment of the appliction ==> continues deployment
* technologies for deployment ==> virtualization and containerization
* managing the prod environment ==> orchestration ==> kubernetes (container level)/ cloud management
* monitor the prod environment ==> continues monitoring
  
Tools like - git, jenkins, ansible, docker, k8s, prometheus, grafana, ansible, etc...

## Process Automation: 
* BPO ==> task automation
* HR Activities ==> new hire, salary processing, quitting formalities
* Business ==> customer onboarding, invoice processing
* Incident management ==> approval workflows
* Leave management ==> approvale workflows
  
Tools like - RPA, BPM, etc...

# Why organizations are using automation

1. To speed up the activities/tasks/process
2. Efficiency (doing teh same task in less time) ==> lowering the operational cost and increases the productivity
3. Accuracy ==> reducing human errors, data/process quality, less mistakes -> less cost of fixing it
4. Compliance ==> ensure process follows standards, maintains the logs, audit trails and reduces the risk o non-compliance penalties
5. Scalability ==> process same acitivites simultaneously on multiple machines, instances, operate efficienty during peak demands, reduce bottlenecks, grow the business to large scales and adapt to the requirments/demands on the go, which will efficently manage the cost

**Benifits in summarized:**
1. Slowness ==> faster
2. Errors ==> Reliable
3. Manual operation ==> Controlled and execute via softwares/machines
4. Limited growth ==> auto scaling and effective managing
5. Cost effective approaches

**Automation tools:**
1. Opensource tools are highly preferred option ==> 1st choice
2. Enterprise solution ==> features and support ==> 2nd choice
