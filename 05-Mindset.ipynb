Shifting from manual exection to automated thinking is less about tools and more about mindset which literally helps to bring change in the system

==> its moving from **doing the task** to **desiging system that does the task**

---
# What Autmated Thinking is all about

Instead of asking:
```
"How do i complete this task"
```
Start asking:
* Can this task be repeated?
* Can i define clear rules for it?
* Can a system do this taskf for me?

---
# Change of Mindset
|Manual Thinking|Automated Thinking|
|---|---|
|Do the task|Desing the workflow for the task|
|Focus on on execution|Focus on repeatibility|
|Solve once|Solve forever|
|Human effort|System-driven execution|

# Automate anything
**Steps involved in automation:**
1. Identify repetative tasks
2. Break the task into standard rules
3. Standard the process
4. Atomate gradually
5. Monitor and Improve
6. Avoid mistakes/errors as much as possible
7. Donot process any taks which is unclear or broken or non manageable
