# Common Tool Misuse & Anti-Patterns

Common AD automation anti-patterns is as follows

# 1. Using Selenium 

* Selenium is a webapplication automation tool
* It cannot support desktop application or api

When Teams automate web UIs (admin portals, legacy tools) using selenium instead of calling API's or native directory tools, it might cause problems

**Why its a problem**
* Fragile: break with any UI changes
* slow and resource-heavy for older applications
* hard to debug or scale esp older application stack
* poor security (credential handling as example)

**Best Approach:**
* use teh API's or native tools to perform the automation


---
# 2. Overloading of Microsoft Active Directory with business logic
What happens Embedding rules like:
* "If department = Finance, assigne these 12 groups"
* "If contractor, expire account in 30 days"
directly insed the AD scripts, group structures or attributes, then it might cause problems

**Why its a problem**
* AD becomes tightly coupled with business rules
* Hard to update when policies changes
* Leads to "mystery logic" which no one will understand
* Complicates audits and compliance

**Best Approach:**
* keep the AD as directory, not as a rules engine
* Move the logic to orchestration tools:
  * ServiceNow
  * ManageEngine
  * Microsoft Power Automate, etc
* Use AD only for
  * identity storage
  * authentication/authorization

---
# 3. Usign the Configuration Tools for approval workflows

What happens:
Trying to force tools like ansible, chef, puppet, terraform to handle
* multi-step approvals
* human decision points
* ticketing logic

**Why it is a problem**
* These tools are build for state management, not for workflows
* leads to awkward, brittle designs
* no proper audit trails for approval

**Best Approach:**
* user workflow/orchestration platforms
  * ServiceNow
  * Azure Logic Apps
* Trigger config tools after approvals are complete

---
# 4. Building automation silos

What happens here, when the different teams build their own:
* scripts
* tools
* process
  
without co-ordination, problems might get triggered
* duplicate logic (eg. multiple onboarding scripts)
* inconsistent outcomes
* difficult to maintain and scale
* no single source of truth

**Best Approach:**
* Centralize the solutions
  * Resuable PowerShell modules
  * shared workflow
* Standardize interfaces (eg. one onboarding pipeline)
* use version control and shared repositories

---
# 5. Ignoring future scalability
Solutions which are build for catering:
* today's needs
* today's volume
* today's process
* manual triggers
* single region enviroments

**Why its a problem:**
* Break under growth
* Break during mergers
* Break during hiring spikes
* Hards to retrofit for automation maturity
* Performance bottlenecks

**Better Approach:**
Design the automation:
* Indempotency (safe re-runs)
* Bulk operations
* Event-driven trigger
* Integration readiness (HR systems, cloud identities)
* module based approaches
* scripting it to support multiple approaches (method overloading, method overriding approaches)
