# Design principles for automation

**Guide to build the automation:**
1. Reliable
2. Scaleable
3. Useful
4. Not fragile
5. Over engineered
6. Not a non sense work

Architecture design will decide whether system will work or break

---
# 1. Understand the entire ecosystem
* All the application features/ modules ==> get the understanding of it
* What your product is
* What are the operational requirement
* Modularize or Module understandings

---
# 2. Simply before you start the design or automation
1. Splitting the functionality or features seperately
2. Eliminating unnecessary tasks, steps, redundancies, bottles
3. Modularize it
4. Optimizing
5. Once it is clear, then only design and then automate


---
# 3. Standardizing the inputs and outputs 
1. Define the data formats
2. Naming Conventions
3. Template related data
4. Reduce variable options
5. Define errors and how to handle it
6. Standardize with all the applicable format

---
# 4. Modularity (Design)
* Break the application into multiple modules as per the application
* Inside each module, define the functionality and its feature properly
* As much as possible, split the methods for resuablity
* Define resuable functions


---
# 5. Handling Exceptions Explicity
1. Dont assume that everything will work or go perfectly
2. Define what can happen when:
   * Data is missing
   * IF the system itself is not available
   * Break of Rule
   * If a rule itself is not applicable
3. Route the exceptions to humans or failling back to some logic
