# Design principles for automation

**Guide to build the automation:**
1. Reliable
2. Scaleable
3. Useful
4. Not fragile
5. Over engineered
6. Not a non sense work

Architecture design will decide whether system will work or break

---
# 1. Understand the entire ecosystem
* All the application features/ modules ==> get the understanding of it
* What your product is
* What are the operational requirement
* Modularize or Module understandings

---
# 2. Simply before you start the design or automation
1. Splitting the functionality or features seperately
2. Eliminating unnecessary tasks, steps, redundancies, bottles
3. Modularize it
4. Optimizing
5. Once it is clear, then only design and then automate


---
# 3. Standardizing the inputs and outputs 
1. Define the data formats
2. Naming Conventions
3. Template related data
4. Reduce variable options
5. Define errors and how to handle it
6. Standardize with all the applicable format

---
# 4. Modularity (Design)
* Break the application into multiple modules as per the application
* Inside each module, define the functionality and its feature properly
* As much as possible, split the methods for resuablity
* Define resuable functions


---
# 5. Handling Exceptions Explicity
1. Dont assume that everything will work or go perfectly
2. Define what can happen when:
   * Data is missing
   * IF the system itself is not available
   * Break of Rule
   * If a rule itself is not applicable
3. Route the exceptions to humans or failling back to some logic


---
# 6. Ensure Transparency and Traceability
* Each step needs to properly documented ==> simple documentation, proper documenting string
* Maintain the actions output like log files, reports, decisions, output/outcomes
* Making us easy to understand, audit, debug
* Process/IT automation ==> releases, compliance & regulations ==> critical

---
# 7. Build For Scalability
* Design the automation script and activities and processs in such a way that, it can be plug and play and it can handle the increased volumes
* Avoid hardcoding, limits, assumptions
* Consider the future scenarios and growth from day one


---
# 8. Prioritize Reliability & Resilince
* include retries (in case of failure)
* if actions are not successful and creates some kind of residues ==> rolled back
* whenever we anticipate an error scenario ==> error handling ==> try/catch
* fail-safe mechanisms should be always implemented
* ensure the sysmtem can recover automatically from interruptions
* Aim for cosistent performance and not just speed

---
# 9. Keep Human in the Loop (HIL) when needed
* Not everything is automatable
* Even if it autmatable, it is advised not to automate 100%
* Allow humans for review
  * Edge casees
  * High risk decisions
  * approval, etc..
* Hybrid systems are often more effective

---
# 10. Design for Maintainability
1. Write clear logic and properly document it
2. Modularize the code
3. Avoid overly complex workflow and keep workflows simple
4. Maintain 1:1 traceability ==> traceability matrix
5. Avoid biased development
6. Never keep for self (donot add dependency)
7. Design it with standards, make sure that your team can easily use it/ understand it with minimal intervention/training
8. It should easily modifiable whenever maintainace activity will be done ==> if possible OOPS

---
# 11. Security and Access Control
1. Protect sensitive data
2. Use proper authentication and permissions
3. Ensure compliance, regulations with data privacy/policies
4. Add environmental files, store in the data vaults, etc ==> avoid hardcoding as much as possible

---
# 12. Measure and Continuously improve
* Track the performance (time saved, error rates, throughput)
* Identify the bottlenecks and refine regularly
* Automation should always revisited which help for continues improvement
* Automation is not **"set and forget"**
