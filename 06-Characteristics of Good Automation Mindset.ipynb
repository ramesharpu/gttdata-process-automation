# Characteristics of Good Automation Mindset

A Strong automation mindset isn't just knowing tools, but its about
1. how you approach to solve the problem
2. processes
3. scale
4. how consistency
5. smart
6. repeated task
7. etc...


---
# 1. Process Oriented Thinking
Think the steps, workflows and systems, not isolated tasks
* Break work into inputs --> action --> output
* See how tasks are connected end to end

---
# 2 Focus on the Repeatability
Identify the repeatable task:
* Ask: ```"Am i doing this again?"``` or ```"is this task repeatable again and againe?"```
* Look for patterns in the activities
* That is ideal candidate for automation

---
# 3. Rule based thinking
Defining clear logic and rules
* If "x" happens --> then do "y"
* Reduce ambiguity in process

Coding/Automation works best when decisions are structured


---
# 4. Standardization
Aim to reduce variations
1. Creating the consistent ways of doing taks
2. Eliminating unncessary exceptions/errors/problems
3. Standardized methodolgies are tried, tested and accepted as the best way to solve the a **particular problem**
4. Foundation for automation ==> as much as possible, we need incorporate as many standards as possible

---
# 5. Continues Improvement (optimatization)
"it works" ==> can i still make it better
* can i make it better
* can i make it faster
* can i make it more effiecient
* can i make it more reliable
* is there a better way to improve or solve this problem
* is there a new workflow which i can take benifit of
* is the existing process no longer used
* etc..

---
# 6. Systems Thinking
1. ISTM, IAM, Infrastructure and business systems are connected
2. Changing in one are, it will affect the other

End to End design which is already there in the system, so be careful while designing it

---
# 7. Risk & Exceptions Awareness
Think about **What could go wrong**
1. Handle the edge cases
2. Handle the negative scenarios
3. Handle all the possible errors
4. Make sure that the automation is not only efficient, but it is also resilient

---
# 8. Tool-Agnostic Approach
Your focus should always be on **logic first and tool second**
1. understand the problem before choosing the tools/ technology
2. avoid over dependece on specific tools


---
# 9. Bias Towards Action
1. Atomate small tasks first
2. Build the system incrementally

---
# 10. Security and Compliance Awareness
You consider access, data and risks
1. Ensure proper permissions
2. Maintain auditability
3. Documentation
4. SOP defined

Especially IAM and Enterprise IT, it is very critical ==> Authentication and Authorization

---
# 10. Collaboration
Never automate in isolation
1. Include the team member in everything
2. Work with stakeholders (IT, HR, Buisness teams)
3. Understand the real need
4. Approvals, Evalution, Feedback, Reviewed
