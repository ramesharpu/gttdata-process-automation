# Hybrid Automation Models
---

### Why Hybrid Atomation is often the right answer

* A single tool rarely covers every aspects of IT operations effectively
* Hybrid automtion combines strengths from multiple systems
  * workflow orchestration
  * identity management
  * infrastructure execution

  ==> into a cohesive approach
* This improves flexibility
* Reduces tool limitations
* allows teams to scale automation without over engineeding with any one of platform

---
# Examples Hybrid Models
#### 1. Workflow + Directory Execution
*ManageEngine(workflow) + Active Directory(execution)*
* Uses a service management or workflow tool to handle approvals and logic
* then exectures changes directly in directory service

#### 2. Indentity Trigger + Infrastructure Automation
*HRM/IAM trigger + Ansible(infrastructure)*
* Events like onboarding or role changes trigger automated infrastructure provisioning

#### 3. Covnersational Interface + Backed Automation
*Bot interface + ManageEngine + Active Directory*
* A chatbot or self service interface initiates requests
* which flow through the orchestration tools
* execute in core systems

**Integration of Powershell/API where ever possible**
- Also include the powershell commands for native workflows ==> speed up the processs
- you can use API requests if its supported
- any kind of permutation and combination with direct API, native application call, web base workflows (selenium), configuration based workflows(ansible)

# Desiging Clean integration between the tools ==> Handoff between multiple tools
* What each product capabilites are
* compare which is the best way to perform an action
  * creating user via powershell command
  * creating user via api call (requests or restassured)
  * creating user via UI (selenium)
  * creating user via ansible
* Finally after the comparision/ evalution, choose the flow which suits the best for your process
* implement it
  * clearly define system ownership (who does what)
  * Use API's or even-based triggers instead of manual flow
  * Pass only necessary information/data between systems
  * implement loggind and traceability across the workflow

# Avoid Duplications and Conflicting Automation
* Centralizing the orchestration
* Write logic in one primary system rather than messing it multiple system
* Modularize and clear define which software is going to used for what activity
* Avoid any kind of overlapping automation
* Set proper expectations
* Set automation rules across the tools
* Write proper product/project and coding documentations
* In the product/project document
  * be clear in the explaination
  * try to give detailed aprroach
  * pref add step by step activities or task
  * add as much examples as possible
  * define the worflows, scenarios, activities
  * prepare some demo videos which the participants or co-workers can use it for reference
  * make sure that it is updated and nothing breaks even over the period of time
* Keep updating/modifying or managing the docs/programs/anything as and when the changes happen/ required
* Establish governance and naming conventions
* Regularly audit workflows to eleminate any redundencies
* Nominate module ownership so that all kinds of disturbance can be reduced to the highest level
