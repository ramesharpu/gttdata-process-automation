# Ansible

---
Ansible is configuration management tool (CMT). 

Ansible automates the management of remote systems and controls their desired state.

---
# Roles of configuration management in automation
* ensures systems are in a desired state, consistent state, across enviroments like on-prem, cloud, hybrid)
* automates provisioning, configuration, and drift correct using declarative definitions
* complements identiy and workflow tools by handling the exectuion layer for infrastructure and application

---
# Strengths
* idempotent automation(safe to run repeatedly without unintended changes ==> if not created, create a new entry, else if it already created do not change)
* strong affinity for infrastructure and system state enforcement (packages, services, configurations)
* agentless architecture
* passwordless authentication should be established via ssh/WinRM
* entire deployment/management of the software/infra will be simplified
* scales well across large fleets of servers
* version-controlled playbooks
* enabling repeatibility adn tracebility using this playbooks

---
# Best fit use cases
* Server Configuration (OS hardening, patch baselines, serivce setups)
* Applciation deployments (consistent installations, updates, rollbacks)
* easy to work across various operating systems with their commands - win, linux, mac => yum, dnf, apt, etc..
* policy enforcements across environment (dev/test/prod parity)
* provisioning and configuration of cloud resources and middleware
* coordinating multi-tier deployments (web/app/db layers)


---
# Limitations
* not suited for approval-driven workflows or human-in-the-loop proces
* limited end-user interactive/self-service capabilities
* not ideal for fine grained identity lifecycle management (approvals, access reviews)
* requires integrations with other systems for event triggers and orchestration

---
# Design considerations
* Maintain clear seperation of concerns
  * workflow/orchestration (ISTM or IAM tools) vs. execution/configuation(ansible)
* keep playbooks modular, reusable, and version controlled
* use inventories (ip or fqdn of vm's) and variables to manage the enviroment differently (very precise, clean)
* integrate with any ci/cd pipelines for controlled deployment and management
* implement logging and reporting for the compliance and troubleshooting
* define handoff eg. workflow tool triggers ansible jobs for execution
* avoid mixing identity automation (user, groups approvals) with infrastructure state unless clearly want it be bounded
