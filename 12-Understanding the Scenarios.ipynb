# Understanding the Scenarios

# Test Scenario
Test Scenario is a n high level description of what needs to validated in a system or process
This defines
* The Workflow to be tested
* The expected outcome
* The conditions under which the tests are going to run

*What are the real-world scenarios that we are trying to test/automate/task*

----
# Different Scenarios
### 1. Business Scenario:
* Representing real-world use cases from a business perspective
* Focus on end-to-end workflows

**Example:**
```
"Customer places an order and receives confirmation"
```


### 2. Functional Scenario:
* Focus on specific system features or functions
* Validate indiviual componenets

**Example:**
```
"System calculates total price correctly"
```

### 3. Positive Scenario and Negative Scenario
**Positive Scenario:**
* Validate expected behavior are under normal conditions
* Examples:
  * Login with valid credentials for successful login
  * Using credit card for payment
  * Using debit card for payment
  * Using netbanking for payment
  * Using UPI for payment
  * Using COD for payment

**Negative Scenario:**
* Test how the system will handle invalid inputs or failures
* Examples:
  * Wrong password --> error message
  * Wrong credit card number
  * Wrong debit card number
  * Wrong netbanking id
  * Pressing refresh button when the payment is process

### 4. Edge cases and boundary conditions
These tests are for checking limits of the systems
* Minimum and maximum values
* Empty or null inputs
* Unexpected user behaviors
* Example:
  * Username field with max characters
  * Clicking on submit or login button with empty fields

### 5. Non Function test cases (Applicable for IT Automation esp. for product development)
* Performance Testing
* Stress Testing
* Load Testing
* Scaleablity Testing
* UI Testing
* DB Testing
* Security Testing
* Peneration Testing
* etc...

# Test Cases are written
1. Based on the scenarios, the testcases are to be written
   * Business/Functional Scenario in mind and formulate different testcases under it
   * Input should be very clear, structured, step by step operation, expected outputs according to ti
2. Whether this testcase is automateable or not


---
# Selecting the right scenarios and testcases for automation
1. Can all the testcases be automation - **May Be**
2. Not everything can be automateable, sometimes it better not to automate
3. Prioritizing the scenarios
   * High Volume and repeatable
   * Stable features (not changing frequently)
   * Critial to business operation
   * Time consuming if done manually
   * Prone to human error
4. Avoid
   * One time test
   * Frequently changing features
   * Highly subjective validations
