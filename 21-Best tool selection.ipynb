# Best Tool Selection

---

# Pref-1: Microsoft Powershell
This is the closest thing to a **"default best tool"**
* Deep, native integration with AD (cmdlets like New-ADUser, Set-ADUser)
* Can automate almost anything in AD (users, groups, permissins, GPO's)
* Highly flexible and scriptable
* Industry standards for AD admins

**Best for:**
* Bulk user provisioning
* Group management
* Scheduled automation
* Complex logic and customization

**Trade-off**
1. Scripting knowledge
2. Approval workflows is not supports


---
# Pref-2: ManageEngine (ADManager Plus)
* Adds approval workflows, delegations, and audit trails
* Automates user lifecycle (JML) with governance
* Reduces manual effor and enforces policies

**Best for:**
* Acess requests
* Approval workflows
* HR driven onboarding/offboarding
* compliance and audit reporting

==> it does not replace powershell, it can orchestrate it(or similar backend actions)

**Trade-off:**
* paid tool
* complex workflows can be overwhelming
* difficult maintain complex operations

---
# pref-3: Ansible
* Great for server configuration and enviroment consistency
* uses playbooks for repeatable automation
* can run powershell under the hood for AD tasks
* directly work with ap plugins

**Best for:**
* managing servers that rely on AD
* enforce system configuration
* large scale enviroment management and consistency can be achieved

**Trade-offs:**
* approval based workflow
* user lifecycle workflows

---
# Pref-4: Selenium
* only work with UI/web UI based operations of AD
* mostly web-only legacy portals
* it can fragile esp with changing uis
* not scalable for larger eco system
